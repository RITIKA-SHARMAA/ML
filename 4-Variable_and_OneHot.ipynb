### Categorical Variables and One Hot Encoding Using pandas

```python
import pandas as pd
```

```python
data = {
    'town': [
        'Delhi', 'Delhi', 'Delhi', 'Delhi', 'Delhi',
        'Shimla', 'Shimla', 'Shimla', 'Shimla',
        'Kolkatta', 'Kolkatta', 'Kolkatta', 'Kolkatta'
    ],
    'area': [2600, 3000, 3200, 3600, 4000, 2600, 2800, 3300, 3600, 2600, 2900, 3100, 3600],
    'price': [550000, 565000, 610000, 680000, 725000, 585000, 615000, 650000, 710000, 575000, 600000, 620000, 695000]
}

df = pd.DataFrame(data)
print(df)
```

**Dummy variables**

```python
dummies = pd.get_dummies(df.town)
# Convert True/False to 0/1
dummies = dummies.astype(int)
dummies
```

```python
#joinging the tables

merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
# droping the town colomn

final = merged.drop(['town'], axis='columns')
final
```

**Dummy Variable Trap**

The Dummy Variable Trap is a scenario that arises when you create dummy variables for a categorical variable and include all the categories as separate columns in your dataset. This can lead to multicollinearity, which occurs when one predictor variable can be linearly predicted from others. Multicollinearity causes issues in models, especially in linear regression, as it makes it difficult to determine the individual effect of each variable.
# New Section

```python
final = final.drop(['Kolkatta'], axis='columns')
final
X = final.drop('price', axis='columns')
X
```

```python
y = final.price
y
```

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
 # need to predict price of house  With :
 # 3400 sqr ft area in Kolkatta
 # 2800 sqr ft home in Shimla

model. predict ([[3400,0,0]])
```

```python
model. predict ( [[2800,0,1]])
```

One Hot Encoding using sklearn

```python
 # First step is to use label encoder to convert town names into numbers

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle

# delhi - 0
# shimla - 1
# Kolkatta - 2
```

```python
X = dfle[['town','area']].values
X # 2D array
```

```python
y = dfle.price.values
y # 1D array
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X # removing 1st colomn
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,3400]])  # 3400 sqr ft area in Kolkatta
```

```python
model.predict([[0,1,2800]])
 # 2800 sqr ft home in Shimla
```
